# Import Libraries

```python
import numpy as np
import pandas as pd
```

# Load data

```python
train_filename = "../data/train.csv"
test_filename = "../data/test.csv"

pre_processed_test_filename = "../data/pre_train.csv"
pre_processed_train_filename = "../data/pre_test.csv"

df1 = pd.read_csv(train_filename)
df2 = pd.read_csv(test_filename)

df = pd.concat([df1, df2])
df.head()
```

## Remove NA values - with mean

```python
df.info()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
df['Embarked'] = df['Embarked'].fillna('Q')
df.info()
```

# Feature Engineering

```python
def substrings_in_string(big_string, substrings):
    big_string = str(big_string)
    for substring in substrings:
        if str.find(big_string, substring) != -1:
            return substring
#     print(big_string)
    return np.nan
```

## Taking title from the people names

```python
title_list=['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev',
                    'Dr', 'Ms', 'Mlle','Col', 'Capt', 'Mme', 'Countess',
                    'Don', 'Jonkheer']

df['Title']=df['Name'].map(lambda x: substrings_in_string(x, title_list))
```

```python
df.head()
```

## Turning cabin number to Deck 

```python
#Turning cabin number into Deck
cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'T', 'G', 'Unknown']
df['Deck']=df['Cabin'].map(lambda x: substrings_in_string(x, cabin_list))
# df.replace({'Deck':{np.nan, 'Unknown'}}, inplace=True)""
df.fillna(value={'Deck': cabin_list[-1]}, inplace=True)
```

```python
df.head()
```

## Taking family size as a new feature

```python
df['Family_Size']=df['SibSp']+df['Parch']
df.head()
```

## Title mapping only to 4 features

```python
df['Title'].unique()
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }

df['Title']=df['Title'].map(lambda x: title_mapping[x])
```

```python
df.info()
```

## Age mapping

```python
def age_map(real_age):
    if real_age < 16:
        return 'child'
    elif real_age < 26:
        return 'young'
    elif real_age < 36:
        return 'middle'
    elif real_age < 62:
        return 'umiddle'
    else:
        return 'old'

df['Age'] = df['Age'].map(lambda x: age_map(x))
```

```python
df.info()
```

## Fair per person

```python
df['Fare_Per_Person']=df['Fare']/(df['Family_Size']+1)
df.head()
```

```python
df.info()
```

```python
df['Is_Alone'] = 0
df.loc[df['Family_Size'] == 1, 'Is_Alone'] = 1
```

## Convert to categorical data to one hot encoding

```python
x_sex = pd.get_dummies(df.Sex, prefix='s')
X_emb = pd.get_dummies(df.Embarked, prefix='emb')
X_age = pd.get_dummies(df.Age, prefix='age')
X_deck = pd.get_dummies(df.Deck, prefix='deck')
X_title = pd.get_dummies(df.Title, prefix='title')
df = pd.concat([df, X_emb, X_deck, X_title, X_age], axis=1)

df.info()
df.drop(['Sex', 'Embarked', 'Deck', 'Title', 'Age', 'Name', 'Cabin', 'Ticket'], axis = 1, inplace=True)
# df.head()
```

```python
train = df[df['Survived'].notnull()]
test = df[df['Survived'].isnull()]
```

```python
train.to_csv(pre_processed_train_filename, index=False, sep=',')
train.info()
```
